A Dynamical System describes the evolution of some state space according to a fixed law. 
It consists of 3 objects: 

1. A phase space $X$, whose elements represent the possible states of the system.
2. The notion of Time, either discrete or continuous.
3. The time evolution law, $\phi : X \to X$



#### What are we Studying?

- The _orbit_ of a point $x \in X$ is the sequence $x, \phi(x), \phi^2(x), \dots, \phi^k(x), \dots$

The theory of dynamical systems focuses on _asymptotic_ behavior of orbits.


Let's take a look at some examples of varying behavior. Consider the family of logistic maps 

<h1><center> $\phi_\theta(x) =  \theta x(1-x)$ </center></h1>


```python
def logistic(theta, x): 
    return theta * x * (1 - x)

xs = np.arange(0,1,0.01)
ys = logistic(2, xs)

plt.plot(xs, ys)
plt.title("logistic map. Theta = 2")
plt.show()
```

```python
def orbit_logistic(theta, x, n):
    p = [x]
    for i in range(n): 
#         x = logistic(theta, x)
        x *= 2
        x -= np.floor(x)
        p.append(x)
    return p
```

```python
def plot_orbits(theta, n):
    all_xs = []
    for x in [0.7]:
        all_xs.append(orbit_logistic(theta, x, n))
    for xs in all_xs:
        plt.plot(range(len(xs)), xs)
```

```python
plot_orbits(0.9, 20)
plt.title("theta = 0.9")
plt.xlabel("# of iterations")
plt.show()
```

```python
plot_orbits(1.5, 20)
plt.title("theta = 1.5")
plt.xlabel("# of iterations")
plt.show()
```

- An orbit is _periodic_ if there exists $p \in \mathbb{N}$ such that $\phi^p(x) = x$. We say it has period $p$.


- An _attractor_ is a subset $A$ of phase space, invariant under $\phi$ (that is, $\phi(A) \subset A$) such that nearby states approach $A$ in the limit. Eg. Limit Cycles

- _chaos_ : Small differences in initial conditions yield widely diverging outcomes
