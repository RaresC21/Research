```python
def make_simplex(p, options  = Options()) : 
    dim = len(p)
    
    A = []
    b = np.zeros(dim+1)
    b[-1] = 1
    
    last = np.zeros(dim)
    for i in range(dim): 
        v = np.zeros(dim)
        v[i] = -1
        
        A.append(v)
        
        last[i] = 1.0 / p[i]
    A.append(last)
    A = np.array(A)
    pt = np.ones(dim) / (dim+1)
    poly = Polytope(A, b, pt, options = options)
    return poly
```

```python
def plot_convergence(p): 
    plt.figure(figsize=(10, 10))
    final = p.all_marginals[-1]
    difference = []
    for m in p.all_marginals:
        diff = 0
        for a,b in zip(m, final): 
            diff += abs(a - b)
        difference.append(diff)
    difference = difference[10:]
    plt.plot(range(len(difference)), difference)
```

```python
plt.figure(figsize=(10, 10))
for j in [50000, 100000, 200000]:
    pts = simplex.pts[:j]
    xxs = []
    bins = np.zeros(50)
    all_ = np.arange(0, 50, 1)
    for i in range(len(pts) - 1): 
        a = pts[i][0]
        b = pts[i + 1][0]
        length = np.linalg.norm(pts[i] - pts[i+1])

        aa = min(a, b)
        bb = max(a, b)
        a = aa 
        b = bb
        for k in range(len(bins) - 1):
            l = k
            r = (k + 1) 
            if b < l or a > r: continue
            rr = min(b, r)
            ll = max(a, l)
    #         print(k, rr - ll)
            bins[k] += (rr - ll) * length / (b - a)
    cur = 0
    for k in range(len(bins)): 
        cur += bins[k]
    #     print(k, cur)
    # print(bins)
    for k in range(len(bins)): 
        bins[k] /= cur
    plt.plot(all_, bins, label = "{} iterations".format(j))
plt.legend()
plt.title("ball reflections for long simplex - distribution up to 50,000 iterations")
```
