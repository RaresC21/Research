```python
%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np

from Polytope import Polytope
from symmetric import *
from far_point import *
from make_cube import *

%matplotlib inline
plt.figure(figsize=(18, 16))
dim = 10
# poly = make_rotated_cube(dim, options = Options())
poly = symmetric(dim, options = Options())

# zero = get_far_point(poly)

lines = poly.reflection_lines(10000)
print("DONE")
```

```python
def separation(lines, left, right): 
#     plt.figure(figsize=(18, 16))
    dimension = len(lines[0][0])
    vec = np.random.uniform(-1, 1, dimension)
    vec /= np.linalg.norm(vec)
    
    total = np.zeros(dimension)
    f = []
    side = []
    for line in lines:
        pt = line[0]
        total += pt
        f.append(np.dot(vec, total) / (len(f) + 1))
        if np.dot(vec, pt) > 0: side.append(1)
        else: side.append(0)   
    return side, f
```

```python
aa = []
bb = []
for i in range(100): 
    side, f = separation(lines, 0, len(lines))


    # this will show us on average the number of reflections needed to cross from one side of 
    # the hyperplane to the other side.
    a = []
    b = []
    cnt = 1
    for i in range(len(side) - 1):    
        cnt += 1
        if side[i] == 1 and side[i + 1] == 0:
            a.append(cnt)
            cnt = 0
        elif side[i] == 0 and side[i + 1] == 1:
            b.append(cnt)
            cnt = 0
    aa.append(np.mean(a))
    bb.append(np.mean(b))
print(np.max(aa), np.max(bb))
print(np.mean(aa), np.std(aa))
print(np.mean(bb), np.std(bb))
```

```python
def plot_lines(lines): 
    plt.figure(figsize=(10, 10))
    left = 0
    right = len(lines)
    xx = 4
    yy = 3
    for i in range(left, right):
        pt1 = np.array([lines[i][0][xx], lines[i][0][yy]])
        pt2 = np.array([lines[i][1][xx], lines[i][1][yy]])
        plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], color = 'blue')
    plt.show()
```
