```python
import matplotlib.pyplot as plt
import numpy as np

from Polytope import Polytope
from symmetric import *
from far_point import *
from make_cube import *

# %matplotlib inline
plt.figure(figsize=(18, 16))
dim = 10
poly = symmetric(dim, dim*4, options = Options())

marginal_direction = np.random.normal(0,1,dim)
poly.fix_marginal_direction(marginal_direction)

lines = poly.reflection_lines(10000, convergence_eps = 1e-6)
```

```python
def plot_convergence(polys, labels): 
    d = []
    for p in polys:
        plt.figure(figsize=(10, 10))
        final = p.all_marginals[-1]
        difference = []
        for m in p.all_marginals:
            diff = 0
            for a,b in zip(m, final): 
                diff += abs(a - b)
            difference.append(diff)
        d.append(difference)
    for cur, l in zip(d, labels): 
        plt.plot(range(len(cur)), cur, label = l)
    plt.legend()
    plt.show()
```

```python
plot_convergence([poly], ['flat'])
```

```python
ball_its = []

A = poly.A 
b = poly.b
b = poly.b
dim = 10
delta = 0.01
radius = 0.1
depth = 0
options = Options(reflection = "sphere", \
                        delta = delta, radius = radius, depth = depth)
ball_poly = Polytope(A, b, np.zeros(dim), options = options)
ball_poly.fix_marginal_direction(marginal_direction)
lines = ball_poly.reflection_lines(10000, convergence_eps = 1e-6)
```

```python
plot_convergence([ball_poly], ['ball'])
```

```python
plt.figure(figsize=(18, 16))
ys = np.arange(0.01,1, 0.01)
for m in poly.all_marginals:
    plt.plot(m, ys)
```
