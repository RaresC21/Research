```python
%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np

from Polytope import Polytope
from symmetric import *
from far_point import *
from make_cube import *

dim = 10
delta = 0.01 / np.sqrt(dim)
radius = 0.01
depth = 0
ball_options = Options(reflection = "sphere", \
                        delta = delta, radius = radius, depth = depth)

# %matplotlib inline
plt.figure(figsize=(18, 16))
poly = make_hyperplanes(dim, options = ball_options)
marginal_direction = np.random.normal(0,1,dim)
poly.fix_marginal_direction(marginal_direction)
```

```python
lines = poly.reflection_lines(50000, convergence_eps = 1e-6)
```

```python
# change initial direction by tiny amount
poly2 = make_hyperplanes(dim, options = ball_options)
poly2.reflection_direction = poly.reflection_direction + np.random.random(0,1/1000,dim)
poly2.reflection_point = np.zeros(dim)
poly.current_average = np.zeros(dim); poly.current_length = 0
```

```python
plt.figure(figsize=(10, 10))
left = 0
right = len(lines)
xx = 4
yy = 5
a = 0
for i in range(left, right - 100, 100):
    pt1 = np.array([lines[i][0][xx], lines[i][0][yy]])
    pt2 = np.array([lines[i+100][1][xx], lines[i+100][1][yy]])
    plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], color = 'black', alpha = a)
    a += 1.0 / ((right - left) / 100)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
xs = [p[4] for p in poly.pts]
plt.hist(xs, bins = 100)
plt.title("ball reflections - frequency of hitting boundary at given x1-coorindate")
```

```python
random_options = Options(reflection = "random", \
                        delta = delta, radius = radius, depth = depth)
random_poly = make_hyperplanes(dim, options = random_options)
```

```python
random_lines = random_poly.reflection_lines(50000, convergence_eps = 1e-6, continue_ = False)
```

```python
plt.figure(figsize=(10, 10))
xs = [p[1] for p in random_poly.pts]
plt.hist(xs, bins = 100)
plt.title("ball reflections - frequency of hitting boundary at given x1-coorindate")
```

```python
plt.figure(figsize=(10, 10))
left = 0
right = len(random_lines)
xx = 2
yy = 3
a = 0
for i in range(left, right - 100, 100):
    pt1 = np.array([random_lines[i][0][xx], random_lines[i][0][yy]])
    pt2 = np.array([random_lines[i+100][0][xx], random_lines[i+100][0][yy]])
    plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], color = 'blue', alpha = a)
    a += 1.0 / ((right - left) / 100)
plt.show()
```
