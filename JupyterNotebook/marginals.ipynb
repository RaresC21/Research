```python
%reload_ext autoreload
%autoreload 2

## What happens for ball reflections?  
A = poly.A 
b = poly.b
dim = 2
delta = 0.01 / np.sqrt(dim)
radius = 0.02
depth = 0
options = Options(reflection = "sphere", \
                        delta = delta, radius = radius, depth = depth)
zero = np.zeros(dim)
ball_poly = []
cur_poly = Polytope(A, b, np.zeros(dim), options = options)
for i in range(1): 
    cur_poly = Polytope(A, b, zero, options = options)
    cur_poly.fix_marginal_direction(marginal_direction)
    _ = cur_poly.reflection_lines(500, convergence_eps = 1e-6)
    ball_poly.append(cur_poly)
```

```python
for p in ball_poly: 
    print(len(p.all_marginals))
    plt.plot(p.all_marginals[-1], ys)
#     plt.plot(p.all_marginals[1], ys)
#     plt.plot(p.all_marginals[10], ys)
```

```python
p = ball_poly[0].pts
num = 300 #len(p)
a = 1
for i in range(0,num,1):
    plt.plot([p[i][0], p[i+1][0]], [p[i][1], p[i+1][1]], color = 'blue', alpha = a)

plt.axis('equal')
plt.show()
```

```python
options = Options(reflection = "random", \
                        delta = delta, radius = radius, depth = depth)
cur_poly = Polytope(A, b, np.zeros(dim), options = options)
# random_far = get_far_point(cur_poly)
random_poly = []
for i in range(1): 
    cur_poly = Polytope(A, b, zero, options = options)
    cur_poly.fix_marginal_direction(marginal_direction)
    _ = cur_poly.reflection_lines(10000, convergence_eps = 1e-6)
    random_poly.append(cur_poly)
```

```python
for p in random_poly: 
    for i in [100,9900]: 
        plt.plot(p.all_marginals[i//100], ys)
```
